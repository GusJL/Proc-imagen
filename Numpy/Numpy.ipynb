```python

import matplotlib.pyplot as plt

#Crear la imagen - Establecer alto y ancho
alto = 200
ancho = 200

# Creando un fondo negro a travez de un array numpy
imagen = np.zeros((alto, ancho, 3), dtype=np.uint8)

# Mostrar la imagen con matplotlib
plt.imshow(imagen)
plt.axis('off')  # desactiva ejes y etiquetas
plt.show()
```

```python
#Estableciendo coordenadas y valores de cada canal
imagen[50:150, 50:150, 0] = 255  # seteo canal rojo = 255 (full)
imagen[50:150, 50:150, 1] = 0    # seteo canal verde = 0
imagen[50:150, 50:150, 2] = 0    # seteo canal azul = 0

# mostrar la imagen luego del cambio
plt.imshow(imagen)
plt.axis('off')
plt.show()
```
